```python
from urllib.parse import quote

# Define o HTML da página de cadastro
HTML_DEMO = """
<!DOCTYPE html>
<html lang="pt-br">

<head>
  <meta charset="UTF-8">
  <title>Demo QA - Formulário</title>
  <style>
    body {
      font-family: Arial, sans-serif;
      background: #f4f6f8;
      padding: 30px;
      margin: 0;
    }

    .layout {
      display: flex;
      gap: 24px;
      align-items: flex-start;
    }

    /* Coluna esquerda: formulário */
    .card {
      flex: 0 0 360px;
      background: white;
      padding: 24px;
      border-radius: 12px;
      box-shadow: 0 2px 12px rgba(0, 0, 0, 0.08);
    }

    h1 {
      margin-top: 0;
      font-size: 24px;
    }

    p {
      color: #444;
    }

    label {
      display: block;
      margin-top: 14px;
      font-weight: bold;
    }

    input {
      width: 100%;
      padding: 10px;
      margin-top: 6px;
      border: 1px solid #cfd8dc;
      border-radius: 8px;
      box-sizing: border-box;
    }

    input:disabled {
      background: #f0f0f0;
      color: #888;
      cursor: not-allowed;
    }

    button {
      margin-top: 18px;
      width: 100%;
      padding: 12px;
      border: none;
      border-radius: 8px;
      cursor: pointer;
      font-size: 16px;
      background: #1565c0;
      color: white;
    }

    button:hover {
      background: #0d47a1;
    }

    #btnCancelar {
      background: #757575;
      margin-top: 8px;
    }

    #btnCancelar:hover {
      background: #616161;
    }

    #resultMsg {
      margin-top: 18px;
      padding: 12px;
      border-radius: 8px;
      display: none;
      font-weight: bold;
    }

    .success {
      display: block !important;
      background: #e8f5e9;
      color: #1b5e20;
    }

    .error {
      display: block !important;
      background: #ffebee;
      color: #b71c1c;
    }

    /* Coluna direita: lista de produtos */
    .product-list {
      flex: 1;
      background: white;
      padding: 24px;
      border-radius: 12px;
      box-shadow: 0 2px 12px rgba(0, 0, 0, 0.08);
      min-height: 200px;
    }

    .product-list h2 {
      margin-top: 0;
      font-size: 20px;
      color: #1565c0;
      border-bottom: 2px solid #e3f2fd;
      padding-bottom: 10px;
    }

    #lista-vazia {
      color: #90a4ae;
      font-style: italic;
    }

    #campoPesquisa {
      width: 100%;
      padding: 8px 12px;
      margin-bottom: 14px;
      border: 1px solid #cfd8dc;
      border-radius: 8px;
      box-sizing: border-box;
      font-size: 14px;
    }

    .produto-item {
      border: 1px solid #e0e0e0;
      border-radius: 8px;
      margin-bottom: 10px;
      overflow: hidden;
      transition: box-shadow 0.2s;
    }

    .produto-item:hover {
      box-shadow: 0 2px 8px rgba(21, 101, 192, 0.15);
    }

    .produto-header {
      display: flex;
      align-items: center;
      padding: 12px 16px;
      background: #f5f5f5;
      gap: 12px;
    }

    .produto-header-click {
      display: flex;
      align-items: center;
      gap: 12px;
      flex: 1;
      cursor: pointer;
      user-select: none;
    }

    .produto-header-click:hover .produto-nome-header {
      color: #1565c0;
    }

    .badge-id {
      background: #1565c0;
      color: white;
      border-radius: 6px;
      padding: 2px 10px;
      font-size: 13px;
      font-weight: bold;
      white-space: nowrap;
    }

    .produto-nome-header {
      flex: 1;
      font-weight: bold;
      font-size: 15px;
    }

    .chevron {
      font-size: 12px;
      color: #90a4ae;
      transition: transform 0.25s;
    }

    .btn-editar {
      margin-top: 0;
      width: auto;
      padding: 4px 12px;
      font-size: 13px;
      background: #ff8f00;
      color: white;
      border: none;
      border-radius: 6px;
      cursor: pointer;
      white-space: nowrap;
    }

    .btn-editar:hover {
      background: #e65100;
    }

    .btn-excluir {
      margin-top: 0;
      width: auto;
      padding: 4px 12px;
      font-size: 13px;
      background: #c62828;
      color: white;
      border: none;
      border-radius: 6px;
      cursor: pointer;
      white-space: nowrap;
    }

    .btn-excluir:hover {
      background: #b71c1c;
    }

    .produto-details {
      display: none;
      padding: 14px 16px;
      background: white;
      border-top: 1px solid #e0e0e0;
    }

    .produto-details table {
      width: 100%;
      border-collapse: collapse;
    }

    .produto-details td {
      padding: 6px 0;
      font-size: 14px;
    }

    .produto-details td:first-child {
      color: #607d8b;
      width: 90px;
      font-weight: bold;
    }

    .produto-item.open .chevron {
      transform: rotate(180deg);
    }

    .produto-item.open .produto-details {
      display: block;
    }
  </style>
</head>

<body>

  <div class="layout">

    <!-- Coluna esquerda: formulário -->
    <div class="card">
      <h1 id="formTitulo">Cadastro de Produtos</h1>
      <p id="formDescricao">Preencha os dados abaixo para cadastrar um novo produto.</p>

      <label for="nome">Nome</label>
      <input id="nome" type="text" placeholder="Digite o nome do produto">

      <label for="id">Id</label>
      <input id="id" type="text" placeholder="Digite o identificador do produto">

      <label for="valor">Valor</label>
      <input id="valor" type="number" placeholder="Digite o valor de venda">

      <label for="estoque">Estoque</label>
      <input id="estoque" type="number" placeholder="Digite o estoque atual">

      <button id="submitBtn" onclick="acaoFormulario()">Enviar</button>
      <button id="btnCancelar" onclick="cancelarEdicao()" style="display:none;">Cancelar</button>

      <div id="resultMsg"></div>
    </div>

    <!-- Coluna direita: lista de produtos -->
    <div class="product-list">
      <h2>Produtos Cadastrados</h2>
      <input id="campoPesquisa" type="text" placeholder="Pesquisar por nome ou ID..." oninput="pesquisar()">
      <p id="lista-vazia">Nenhum produto cadastrado ainda.</p>
      <div id="produtos-container"></div>
    </div>

  </div>

  <script>
    const idsCadastrados = [];
    const produtos = [];
    let modoEdicao = false;
    let idEmEdicao = null;

    // ── Decide entre cadastrar ou salvar edição ──
    function acaoFormulario() {
      if (modoEdicao) {
        salvarEdicao();
      } else {
        validarFormulario();
      }
    }

    // ── Cadastro ──
    function validarFormulario() {
      const nome = document.getElementById('nome').value.trim();
      const id = document.getElementById('id').value.trim();
      const valor = document.getElementById('valor').value.trim();
      const estoque = document.getElementById('estoque').value.trim();
      const msg = document.getElementById('resultMsg');

      msg.className = '';
      msg.style.display = 'none';

      if (!nome || !id || !valor || !estoque) {
        msg.textContent = 'Erro: todos os campos são obrigatórios.';
        msg.classList.add('error');
        return;
      }

      if (idsCadastrados.includes(id)) {
        msg.textContent = `Erro: o ID "${id}" já está cadastrado.`;
        msg.classList.add('error');
        return;
      }

      if (Number(valor) <= 0) {
        msg.textContent = 'Erro: o valor deve ser maior que zero.';
        msg.classList.add('error');
        return;
      }

      if (Number(estoque) < 0) {
        msg.textContent = 'Erro: o estoque não pode ser negativo.';
        msg.classList.add('error');
        return;
      }

      idsCadastrados.push(id);
      const p = { id, nome, valor: Number(valor), estoque: parseInt(estoque) };
      produtos.push(p);

      msg.textContent = `Sucesso: registro criado para ${nome} (ID: ${id}) no valor de R$ ${p.valor.toFixed(2)} com o estoque de ${p.estoque}.`;
      msg.classList.add('success');

      adicionarProdutoNaLista(p);

      document.getElementById('nome').value = '';
      document.getElementById('id').value = '';
      document.getElementById('valor').value = '';
      document.getElementById('estoque').value = '';
    }

    // ── Ativa modo edição ──
    function editarProduto(id) {
      const produto = produtos.find(p => p.id === id);
      if (!produto) return;

      modoEdicao = true;
      idEmEdicao = id;

      document.getElementById('formTitulo').textContent = 'Editar Produto';
      document.getElementById('formDescricao').textContent = 'Altere os campos desejados e clique em Salvar.';

      document.getElementById('nome').value = produto.nome;
      document.getElementById('id').value = produto.id;
      document.getElementById('id').disabled = true;
      document.getElementById('valor').value = produto.valor;
      document.getElementById('estoque').value = produto.estoque;

      document.getElementById('submitBtn').textContent = 'Salvar';
      document.getElementById('btnCancelar').style.display = 'block';

      const msg = document.getElementById('resultMsg');
      msg.className = '';
      msg.style.display = 'none';
    }

    // ── Salva a edição após confirmação ──
    function salvarEdicao() {
      const nome = document.getElementById('nome').value.trim();
      const valor = document.getElementById('valor').value.trim();
      const estoque = document.getElementById('estoque').value.trim();
      const msg = document.getElementById('resultMsg');

      msg.className = '';
      msg.style.display = 'none';

      if (!nome || !valor || !estoque) {
        msg.textContent = 'Erro: todos os campos são obrigatórios.';
        msg.classList.add('error');
        return;
      }

      if (Number(valor) <= 0) {
        msg.textContent = 'Erro: o valor deve ser maior que zero.';
        msg.classList.add('error');
        return;
      }

      if (Number(estoque) < 0) {
        msg.textContent = 'Erro: o estoque não pode ser negativo.';
        msg.classList.add('error');
        return;
      }

      const confirmar = confirm(`Deseja realmente salvar as alterações do produto "${nome}" (ID: ${idEmEdicao})?`);
      if (!confirmar) return;

      // Atualiza o objeto no array
      const produto = produtos.find(p => p.id === idEmEdicao);
      produto.nome = nome;
      produto.valor = Number(valor);
      produto.estoque = parseInt(estoque);

      // Atualiza o card na lista da direita
      atualizarCardNaLista(produto);

      msg.textContent = `Produto "${nome}" (ID: ${idEmEdicao}) atualizado com sucesso.`;
      msg.classList.add('success');

      cancelarEdicao();
    }

    // ── Cancela e volta ao modo cadastro ──
    function cancelarEdicao() {
      modoEdicao = false;
      idEmEdicao = null;

      document.getElementById('formTitulo').textContent = 'Cadastro de Produtos';
      document.getElementById('formDescricao').textContent = 'Preencha os dados abaixo para cadastrar um novo produto.';

      document.getElementById('nome').value = '';
      document.getElementById('id').value = '';
      document.getElementById('id').disabled = false;
      document.getElementById('valor').value = '';
      document.getElementById('estoque').value = '';

      document.getElementById('submitBtn').textContent = 'Enviar';
      document.getElementById('btnCancelar').style.display = 'none';
    }

    // ── Adiciona produto na lista ──
    function adicionarProdutoNaLista(produto) {
      document.getElementById('lista-vazia').style.display = 'none';
      const container = document.getElementById('produtos-container');

      const item = document.createElement('div');
      item.className = 'produto-item';
      item.id = `produto-${produto.id}`;

      item.innerHTML = gerarHTMLItem(produto);
      container.appendChild(item);
    }

    // ── Atualiza card existente ──
    function atualizarCardNaLista(produto) {
      const item = document.getElementById(`produto-${produto.id}`);
      if (!item) return;
      const wasOpen = item.classList.contains('open');
      item.innerHTML = gerarHTMLItem(produto);
      if (wasOpen) item.classList.add('open');
    }

    // ── Gera o HTML do card de produto ──
    function gerarHTMLItem(produto) {
      return `
        <div class="produto-header">
          <div class="produto-header-click" onclick="toggleProduto('${produto.id}')">
            <span class="badge-id"># ${produto.id}</span>
            <span class="produto-nome-header" id="nome-display-${produto.id}">${produto.nome}</span>
            <span class="chevron">&#9660;</span>
          </div>
          <button class="btn-editar" onclick="editarProduto('${produto.id}')">Editar</button>
          <button class="btn-excluir" onclick="excluirProduto('${produto.id}')">Excluir</button>
        </div>
        <div class="produto-details" id="details-${produto.id}">
          <table>
            <tr><td>Nome</td><td id="td-nome-${produto.id}">${produto.nome}</td></tr>
            <tr><td>ID</td><td>${produto.id}</td></tr>
            <tr><td>Valor</td><td id="td-valor-${produto.id}">R$ ${produto.valor.toFixed(2)}</td></tr>
            <tr><td>Estoque</td><td id="td-estoque-${produto.id}">${produto.estoque} un.</td></tr>
          </table>
        </div>
      `;
    }

    function excluirProduto(id) {
      var produto = null;
      for (var i = 0; i < produtos.length; i++) {
        if (produtos[i].id === id) { produto = produtos[i]; break; }
      }
      if (!produto) return;

      var confirmar = confirm('Deseja realmente excluir o produto "' + produto.nome + '" (ID: ' + id + ')?');
      if (!confirmar) return;

      // Remove do array de produtos
      for (var j = 0; j < produtos.length; j++) {
        if (produtos[j].id === id) { produtos.splice(j, 1); break; }
      }

      // Remove do array de IDs cadastrados
      var idxCad = idsCadastrados.indexOf(id);
      if (idxCad !== -1) idsCadastrados.splice(idxCad, 1);

      // Remove o card do DOM
      var card = document.getElementById('produto-' + id);
      if (card && card.parentNode) {
        card.parentNode.removeChild(card);
      }

      // Mostra aviso de lista vazia se necessário
      if (produtos.length === 0) {
        document.getElementById('lista-vazia').style.display = '';
      }
    }

    function pesquisar() {
      var termo = document.getElementById('campoPesquisa').value.toLowerCase().trim();
      var cards = document.querySelectorAll('.produto-item');
      var visiveis = 0;
      for (var i = 0; i < cards.length; i++) {
        var card = cards[i];
        var idProd = card.id.replace('produto-', '');
        var nomeProd = card.querySelector('.produto-nome-header').textContent.toLowerCase();
        if (idProd.indexOf(termo) !== -1 || nomeProd.indexOf(termo) !== -1) {
          card.style.display = '';
          visiveis++;
        } else {
          card.style.display = 'none';
        }
      }
      document.getElementById('lista-vazia').style.display =
        (visiveis === 0 && produtos.length > 0) ? '' : (produtos.length === 0 ? '' : 'none');
    }

    function toggleProduto(id) {
      const item = document.getElementById(`produto-${id}`);
      item.classList.toggle('open');
    }
  </script>
</body>

</html>-
"""

# Gera a URL para o Selenium acessar
DEMO_URL = "data:text/html;charset=utf-8," + quote(HTML_DEMO)
print("Página HTML preparada com sucesso.")
```

```python
# Configura o tempo de espera e carrega a página
wait = WebDriverWait(driver, 10)
driver.get(DEMO_URL)

# Localiza os elementos da tela
campo_nome = wait.until(EC.presence_of_element_located((By.ID, "nome")))
campo_id = driver.find_element(By.ID, "id")
campo_valor = driver.find_element(By.ID, "valor")
campo_estoque = driver.find_element(By.ID, "estoque")
botao_enviar = driver.find_element(By.ID, "submitBtn")

# Preenche o formulário com dados válidos
campo_nome.clear()
campo_nome.send_keys("Areia pa kh")

campo_id.clear()
campo_id.send_keys("7")

campo_valor.clear()
campo_valor.send_keys("19.90")

campo_estoque.clear()
campo_estoque.send_keys("500")

# Executa a ação e espera o feedback visual
botao_enviar.click()

mensagem_elemento = wait.until(
    EC.visibility_of_element_located((By.ID, "resultMsg"))
)

# Valida se o resultado foi o esperado (Assertion)
mensagem = mensagem_elemento.text
print("Mensagem retornada:", mensagem)

assert "Sucesso" in mensagem
assert "Areia pa kh" in mensagem

print("Teste positivo executado com sucesso.")
```

```python
!pip install -q selenium

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

def novo_driver():
    options = Options()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    return webdriver.Chrome(options=options)
```

```python
driver = novo_driver()
driver.get(DEMO_URL)
time.sleep(1)

driver.find_element(By.ID, "nome").send_keys("Produto A")
driver.find_element(By.ID, "id").send_keys("111")
driver.find_element(By.ID, "valor").send_keys("10")
driver.find_element(By.ID, "estoque").send_keys("5")
driver.find_element(By.ID, "submitBtn").click()
time.sleep(1)

driver.find_element(By.ID, "nome").send_keys("Produto B")
driver.find_element(By.ID, "id").send_keys("222")
driver.find_element(By.ID, "valor").send_keys("20")
driver.find_element(By.ID, "estoque").send_keys("8")
driver.find_element(By.ID, "submitBtn").click()
time.sleep(1)

pesquisa = driver.find_element(By.ID, "campoPesquisa")
pesquisa.send_keys("Produto B")
time.sleep(1)

cards = driver.find_elements(By.CSS_SELECTOR, ".produto-item")
visiveis = [card for card in cards if card.is_displayed()]

assert len(visiveis) == 1
assert "Produto B" in visiveis[0].text

print("Pesquisa Realizada")
driver.quit()
```
